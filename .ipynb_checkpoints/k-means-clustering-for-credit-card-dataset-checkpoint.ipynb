# Data Cleaning and Preprocessing

### Handling missing values

```python
# Check for missing values
df.isnull().sum()
```

```python
# Fill null values by using median
df = df.drop('CUST_ID', axis=1)
df.fillna(df.median(), inplace=True)
df.isnull().sum()
```

Missing values handled

__In clustering analysis, the objective is not to perform predictions but to identify and group observations that exhibit similar patterns. Therefore, clustering does not involve a dependent (target) variable to be predicted.__

```python
# Handle skewness
df_log = df.copy()
for col in df.columns:
    df_log[col] = np.log1p(df[col])
```

```python
# Remove Outliers
def remove_outliers(df):
    Q1 = df.quantile(0.25)
    Q3 = df.quantile(0.75)
    IQR = Q3 - Q1
    
    return df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]

df = remove_outliers(df_log)
```

```python
# Perform Feature Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.9)  # keep 90% variance
X_pca = pca.fit_transform(X_scaled)

print("Number of components:", pca.n_components_)
```

### Elbow Method

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score

wcss = []
silhouette_scores = []
db_scores = []

K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_pca)
    
    wcss.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_pca, labels))
    db_scores.append(davies_bouldin_score(X_pca, labels))
```

```python
# Plot Evaluation Metrics
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.plot(K_range, wcss, marker='o')
plt.title('Elbow Method')

plt.subplot(1,3,2)
plt.plot(K_range, silhouette_scores, marker='o')
plt.title('Silhouette Score')

plt.subplot(1,3,3)
plt.plot(K_range, db_scores, marker='o')
plt.title('Davies-Bouldin Score')

plt.show()
```

```python
optimal_k = 4  # choose based on plots

kmeans = KMeans(n_clusters=optimal_k, random_state=42)
labels = kmeans.fit_predict(X_pca)

df['Cluster'] = labels
```

```python
pca_2d = PCA(n_components=2)
X_vis = pca_2d.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
plt.scatter(X_vis[:,0], X_vis[:,1], c=labels, cmap='viridis')
plt.title("Clusters Visualization (PCA 2D)")
plt.show()
```

```python
cluster_profile = df.groupby('Cluster').mean()
print(cluster_profile)
```

```python
def label_cluster(row):
    if row['PURCHASES'] > 1000:
        return "High Spender"
    elif row['BALANCE'] > 3000:
        return "Debt Heavy"
    else:
        return "Low Activity"

df['Segment'] = df.apply(label_cluster, axis=1)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=2, min_samples=5)
db_labels = dbscan.fit_predict(X_pca)

print("Unique clusters:", set(db_labels))
```

```python
# Perform Standardization
from sklearn.preprocessing import StandardScaler

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

## Elbow Method
Plot the Within-Cluster Sum of Squares (WCSS) as a function of the number of clusters (k). The optimal number of clusters is identified at the point where the curve begins to level off, known as the elbow.

```python
from sklearn.cluster import KMeans

wcss_list = []
for i in range(1, 11):
    model = KMeans(n_clusters=i, random_state =42)
    model.fit(X_scaled)
    wcss_list.append(model.inertia_) # wcss value
    #print(model.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss_list, marker='o')  
plt.title('The Elbow Method Graph')  
plt.xlabel('Number of clusters(k)')  
plt.ylabel('wcss_list')  
plt.show()
```

```python
# Performing predictions when k=5
kmeans = KMeans(n_clusters=5, random_state= 42)  
y_predict= kmeans.fit_predict(X_scaled)  
print(y_predict) #Predicted labels
```

```python
df['cluster']=y_predict
print(df)
```

```python
print(kmeans.cluster_centers_)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.scatter(X_pca[:, 0],X_pca[:, 1],c=y_predict,cmap='rainbow',alpha=0.7)

centers_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(centers_pca[:, 0],centers_pca[:, 1],marker='X',s=200)
plt.title('K-Means Clusters (PCA - 2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total variance explained by 3 PCs:", sum(pca.explained_variance_ratio_))
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(X_scaled, kmeans.labels_)
print("Silhouette Score:", score)
```

__The obtained Silhouette score indicates that the clusters are not well separated, which is consistent with the observations from the above 2D PCA plot. In general, a higher Silhouette score, closer to 1, indicates well-separated and clearly defined clusters.__
